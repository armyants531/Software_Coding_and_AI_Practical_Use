프로젝트 주제 : 남방톱날꽃게(청게)의 나이 예측


주제 선정 이유 및 목표 :

간장게장을 즐겨 먹는 나는 평소 게에 대한 관심이 많았다. 그런데 게를 보고 대략적인 나이를 예측할 수 있을까?
게는 죽을 때까지 탈피를 거듭하며 몸집을 키워나간다. 그래서 크기나 무게등을 이용하여 게의 나이를 예측할 수 있을 것이라는 생각이 들었다. 교수님께서 추천해주신 Kaggle 사이트에 들어가 crab을 검색해보니 Crab Age Prediction 데이터셋을 찾을 수 있었고 남방톱날꽃게(청게)의 나이 예측을 주제로 선정하게 되었다.   
나의 목표는 다중 선형 회귀 분석을 통해 꽃게의 몸무게, 높이, 둘레 등의 정보를 입력받으면 나이를 예측하는 딥러닝 모델을 만드는 것이다. 정확한 모델을 찾기 위해 홀드 아웃 교차 검증 뿐만 아니라 수업 시간에 다루었던 K-겹 교차 검증, EarlyStopping() 함수 등 사용해 보고자 한다. 또한, 다양한 시각화 방식을 선택하여 데이터 분석을 해 볼 생각이고, tensor bored를 이용해 kernel이나 bias를 확인해보고자 한다.  


과제 수행하면서 느낀점 : 

 맨 처음 부딪힌 난관은 주제 선정이었다. 유정란과 무정란 분류, 썩은 고기 판별 등등 하고 싶었던 주제는 많았지만, 이를 수치화된 데이터셋으로 잘 정리한 것을 찾기가 힘들었기 때문이다. 처음에는 레드 킹크랩과 블루 킹크랩의 분류에 대한 연구를 하고 싶었지만, Kaggle이나 구글 검색으로는 데이터셋을 찾을 수가 없었다. 하지만 이후에 crab이라는 더 넓은 범위의 단어를 검색하니 잘 수치화된 데이터셋을 찾을 수 있었다. 이러한 시행 착오를 거치면서 데이터를 가지고 좋은 딥러닝 모델을 만드는 것도 중요하지만, 일상생활에서 데이터를 수집하는 일은 더욱 중요하다는 생각이 들었다. 데이터가 없으면 딥러닝 모델을 만들 수 없기 때문이다.      
 한편, 본 과제를 수행하기 위해 수업 자료 뿐만 아니라 인터넷 사이트에 들어가서 정보를 얻기도 하고 인프런 딥러닝 강좌를 들으면서 지식을 쌓기도 하였다. 데이터 가공 과정을 두 번에 걸쳐서 해야한다는 말을 듣고, 나는 처음에는 '굳이 2개의 파일에 나누어서 전처리를 해야 할까?'라는 생각이 들었지만 데이터 전처리에 대한 공부를 하고 난 이후에는 결측치, 중복값, 이상치 제거 등 해야할 것이 많다는 것을 알게 되었고 이전보다 많은 지식을 얻었다는 생각에 뿌듯함을 느꼈다. 또한 데이터를 분석하고 시각화하는 과정에서 많은 방법을 사용하였는데 수업시간에 배웠던 내용을 복습하고 인터넷 검색을 해보면서 내가 가진 데이터에 대한 이해도를 높일 수 있었다.   
 하지만 역시 가장 큰 난관이자 나의 발전에 가장 큰 도움을 주었던 활동은 딥러닝 분석이다. 수업 시간에 들었던 딥러닝 분석 예시에는 다중 선형 회귀 분석이 없어 스스로 공부를 해야하는 부분이 꽤 많았었다. 예를 들면 딥러닝 모델을 compile 할 때 loss(손실 함수), optimizer(정규화기), metrics(평가지표)는 어떻게 설정하는지 같은 부분이었다. 이러한 활동을 통해 스스로 공부할 수 있는 능력을 많이 기를 수 있었던 것 같다. 이번 학기에 내가 소프트웨어 코딩과 AI 활용이라는 과목을 들은 목적은 인공지능 연구원이라는 내 꿈에 한 층 더 가까워지기 위함이었는데 이번 딥러닝 프로젝트가 막연했던 나의 꿈을 구체화시켜준 것 같다. 이제 곧 훈련소에 입대하여 군생활을 하게 되는데 자대 배치 이후에 시간이 된다면 딥러닝 분석에 대한 공부를 계속하고 좋은 아이디어가 떠오른다면 개인 프로젝트를 진행해보고 싶다.
